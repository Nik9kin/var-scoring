```python
import networkx as nx
import csv
```

```python
with open('lib.info.norine') as lib:
    with open('norine_spectrums.mgf') as f1:
        with open('norine_spectrums_ids.mgf', 'w') as f2:
            for line in f1.readlines():
                f2.write(line[:-1])
                if line == 'TITLE=\n':
                    pep_id = int(lib.readline()[11:16])
                    f2.write(str(pep_id))
                f2.write('\n')
```

```python
norine_gnps_cluster = nx.Graph()
with open('clustered_spectra/norine_edges.txt') as f:
    for v in f.readlines():
        norine_gnps_cluster.add_edge(int(v.split()[0]), int(v.split()[1]))
print(norine_gnps_cluster.number_of_nodes(), norine_gnps_cluster.number_of_edges())
```

```python
ccs = list(nx.connected_components(norine_gnps_cluster))
ccs = [list(cc) for cc in ccs]
ccs.sort()
ccs
```

```python
with open('clustered_spectra/norine_clusters.txt', 'w') as f:
    for cc in ccs:
        f.write(' '.join([str(i) for i in cc])+'\n')
```

```python
import pandas as pd


data = pd.read_csv('norineText_short.csv')
data.head()
```

```python
norine_families = {}
with open('norineText_short.csv') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        row['id_peptide'] = int(row['id_peptide'][3:])
        if norine_families.get(row['family']) == None:
            norine_families[row['family']] = [row['id_peptide']]
        else:
            norine_families[row['family']].append(row['id_peptide'])
norine_families.pop('other')
norine_families.pop('unknown')
norine_families
```

```python
print('families with greater then 20 pnps:')
for name, ids in norine_families.items():
    if len(ids) > 20:
        print(name)
print()
print('families with only 1 pnp:')
for name, ids in norine_families.items():
    if len(ids) == 1:
        print(name)
```

```python
with open('norine_families.txt', 'w') as f:
    for name, ids in norine_families.items():
        f.write(' '.join([str(i) for i in ids]) + '\n')
```
